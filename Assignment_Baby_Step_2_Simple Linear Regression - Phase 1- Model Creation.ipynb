```python
# Pandas is a library which is used for data manipulation and analysis.
import pandas as pd # 'import' has memory consumtion which is compare to 'from' method 
```

```python
# Data Collection 
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset # Visualize the dataset
```

```python
# Split Input and Output
independent = dataset[["YearsExperience"]] # Assigning YearsExperience column as an input and storing in the independent variable 
```

```python
independent # Visualize the independent variable
```

```python
dependent = dataset[["Salary"]] # Assigning Salary column as a output and storing in the dependent variable 
dependent # Visualize the dependent varaiable 
```

```python
# Split train set and test set
from sklearn.model_selection import train_test_split # import the train_test_split function from sklearn.model_selection
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #Assigning train set input,test set input, train set output and test set output respect to X_train,X_test,y_train,y_test 
```

```python
# Learning Phase (Model Creation)
from sklearn.linear_model import LinearRegression # import LinearRegression Function from sklearn.linear_model
regressor = LinearRegression() # Assigning LinearRegression function to regressor variable 
regressor.fit(X_train,y_train) # Adjusting its internal parameters so it can map the input features X_train to the target values y_train.
```

```python
weight = regressor.coef_ # Assigning the slope value to the weight variable
weight
```

```python
bias = regressor.intercept_ # Assigning the intercept value to the bias variable
bias
```

```python
# Evaluation
y_pred = regressor.predict(X_test) # Uses the trained regressor model to generate predicted target values (y_pred)
```

```python
# Evaluation Metrics
from sklearn.metrics import r2_score  # import r2_score function from sklearn.metrics
r_score = r2_score(y_test,y_pred) # Finding the r2_score value and assigning to the r_score variable
```

```python
r_score # Displaying the r_score to check whether it is near to the value 1. So it will be a good model.
```

```python
# Save the model
import pickle # Loads the pickle module
filename = "finalized_model_linear.sav" # holding the desired file name (and extension) where the serialized model will be stored or retrieved
```

```python
pickle.dump(regressor,open(filename,'wb')) # Serializes the trained regressor object and writes it to the file specified by filename in binary mode
```

```python
# Load the model
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))
result = loaded_model.predict([[15]]) # Feeds the single input value 15 to predict the output
```
